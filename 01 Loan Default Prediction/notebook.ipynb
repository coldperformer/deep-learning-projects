# **1. Introduction**
---

- Optimization is the **selection** of the **best element** from some set of **available alternatives**.

- It is the most **essential ingredient** in the recipe of machine learning algorithms.

- It initiates with defining some sort of **loss/cost function** and ends with minimizing it.

- The choice of an optimized algorithm can **make** a **difference** between **getting good accuracy** in hours or days.

## **1.1 Gradient Descent**

- It is a **generic optimization algorithm** capable of finding optimal solutions to a wide range of problems.

- Here the general idea is to **tweak parameters iteratively** to **minimize** a **cost function**.

- **For example**, you are lost in the mountains in a dense fog, and you can only feel the slope of the ground below your feet.

- A good strategy to get to the bottom of the valley quickly is to go downhill in the direction of the steepest slope.

- It **measures** the **local gradient** of the **error function** in the **direction of descending** concerning the parameter vector θ.

- Once the **gradient** is **zero**, you have reached a **minimum**!

- **Working:**
  - You start by filling θ with random values, also called **random initialization**.

  - Then you **improve** it **gradually**, taking one baby step at a time.

  - Each step attempts to **decrease** the **cost function** (e.g., the MSE) until the algorithm converges to a minimum.
</br>

### **1.1.1 Batch Gradient Descent**

- In this variant, **calculations** are **involved** over the **full training set** X i.e. at each Gradient Descent step.

- As a result, it is **terribly slow** on very **large training sets**.

- However, Gradient Descent **scales well** with the **number of features**.

- **For example**, let's say there are hundreds of thousands of features, 

- Training a Linear Regression model using **Gradient Descent** is **faster** than using the **Normal Equation**.

<center><img src="https://www.xpertup.com/wp-content/uploads/2018/05/1-1.gif"></center>

<a name=Section112 id=Section112></a>

### **1.1.2 Stochastic Gradient Descent**

- This variant **picks** a **random instance** in the training set at **every step** and **computes** the **gradients** based only on that **single instance**.

- Working on a single instance at a time makes the algorithm much **faster** because it has very **little data** to **manipulate** at every **iteration**.

- It also makes it **possible** to **train** on **huge training sets**, since only one instance needs to be in memory at each iteration.

- This algorithm **instead** of **gently decreasing** until it reaches the minimum, the **cost function** will **bounce up and down**.

- Over time it will end up very **close to the minimum**, but once it gets there it will **continue to bounce around**, never settling down.

- When using Stochastic Gradient Descent, the training **instances** must be **independent** and **identically distributed** (IID).

- It ensures that the parameters get pulled toward the global optimum, on average.

</br>
<center><img width="50%" src="https://raw.githubusercontent.com/insaid2018/Term-4/master/images/stochastic%20gradient%20descent.png"></center>

<a name=Section113 id=Section113></a>

### **1.1.3 Mini-Batch Gradient Descent**

- In this variant, **gradients** are **computed** on **small random sets** of instances called **mini-batches**.

- It gives a **performance boost** from **hardware optimization** of matrix operations, especially when using GPUs.

- As a result, it will **end up walking around** a **bit closer** to the **minimum** than Stochastic GD.

- It is **less erratic** than with Stochastic GD, especially with fairly large mini-batches.

- It **requires** a **good learning schedule** to reach the minimum.

</br>
<center><img width=65% src="https://raw.githubusercontent.com/insaid2018/Term-4/master/images/mini-batch-gradient-descent.png"></center>

<a name=Section114 id=Section114></a>

### **1.1.4 Issues with Gradient Descent**

- If the learning rate is too small, then the algorithm will have to go through many iterations to converge, which will take a long time.

</br>
<center><img width="65%" src="https://raw.githubusercontent.com/insaid2018/Term-4/master/images/gradient%20con%201.png"></center>
</br>

- If the learning rate is too high, you might jump across the valley and end up on the other side, possibly even higher up than before.

</br>
<center><img width="65%" src="https://raw.githubusercontent.com/insaid2018/Term-4/master/images/gradient%20con%202.png"></center>
</br>

- Not all cost functions look like nice, regular bowls. There may be holes, ridges, plateaus, and all sorts of irregular terrains.

  - This makes convergence to the minimum difficult.

  - If the **random initialization** starts on the **left**, then it will **converge** to a **local minimum** (not as good as the global minimum).
  
  - If it **starts** on the **right**, then it will take a very **long time** to **cross** the **plateau**. 
  - And if you **stop too early**, you will **never reach** the **global minimum**.

</br>

**<h4>KNN Imputer:</h4>** 

- This is an **effective approach** where data **imputation** is **done** using **prediction** by a model.

- This approach is only effective for numerical data.

- A **model** is **created** for **each feature** that has **missing values**, taking as input values of all other input features.

- A **new sample** is **imputed** by **finding** the **samples** in the training set **“closest”** to it and **averages** these **nearby points** to **fill in the value**.

**<h4>K Fold Target Encoding:</h4>**

- **Target encoding** is **one of the most powerful** **techniques** in feature engineering.

- It has been **widely applied** and **developed** in **different forms**.

- The **limitation** of **Target Encoding** is the **overfitting** of the **data**.

- The **goal** of **K Fold Target Encoding** is to **reduce** the **overfitting** by **adding regularization** to the mean encoding.

- Let's take an example as follows:

|Example|Calculation |
|:-------------------------:|:-------------------------:|
|![img1](https://raw.githubusercontent.com/insaid2018/Domain_Case_Studies/master/Telecom/te1.PNG)  | Here the count of A = 10, B = 10. The Mean(A): 6/10 = 0.6, Mean(B): 3/10 = 0.3|

- In the above diagram, we can observer overfitting over label A. To handle this overfitting we will perform 5 fold target encoding. 

- Each fold's categories are encoded based on the mean of the rest of the fold's categories.

- For first fold, mean values are estimated based on the rest of the folds i.e. Fold 2, Fold 3, Fold 4 & Fold 5.

```
class KFoldTargetEncoder(BaseEstimator, TransformerMixin):
  def __init__(self ,colnames , targetName, n_fold = 5, verbosity = True, discardOriginal_col = False):
    self.colnames = colnames
    self.targetName = targetName
    self.n_fold = n_fold
    self.verbosity = verbosity
    self.discardOriginal_col = discardOriginal_col

  def fit(self, X, y = None):
    return self

  def transform(self, X):
    assert(type(self.targetName) == str)
    assert(type(self.colnames) == str)
    assert(self.colnames in X.columns)
    assert(self.targetName in X.columns)
    mean_of_target = X[self.targetName].mean()
    kf = KFold(n_splits = self.n_fold, shuffle = False, random_state = 42)
    col_mean_name = self.colnames + '_' + 'Enc'
    X[col_mean_name] = np.nan
    for tr_ind, val_ind in kf.split(X):
      X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
      X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(X_tr.groupby(self.colnames)[self.targetName].mean())
      X[col_mean_name].fillna(mean_of_target, inplace = True)
    if self.verbosity:
      encoded_feature = X[col_mean_name].values
      print('Correlation between the new feature, {} and, {} is {}.'.format(col_mean_name, self.targetName, np.corrcoef(X[self.targetName].values, encoded_feature)[0][1]))
    if self.discardOriginal_col:
      X = X.drop(self.colnames, axis=1)
    return X
```

## **8.2 Batch Gradient Descent**

- Here we will implement gradient descent from scratch as there is **no official implementation available**.

- There are a variety of algorithms that are **far more powerful** than the implementation shown below.

- We recommend you take its working as a high-level overview of how it works in real life.

```
class GradientDescent(BaseEstimator, ClassifierMixin):
  def __init__(self ,X, y, random_state=None, epochs=1000, learnrate=0.1):
    self.X = X
    self.y = y
    self.random_state=random_state
    self.epochs=epochs
    self.n_records, self.n_features = self.X.shape 
    self.last_loss = None
    self.learnrate = learnrate
    self.weights = np.random.normal(scale=1 / self.n_features**.5, size=self.n_features)

  def sigmoid(x):
    """Calculate sigmoid"""
    return 1/(1+np.exp(-x))

  def fit(self, X, y):

    self.classes_ = np.unique(y)

    # Use to same seed to make debugging easier
    np.random.seed(self.random_state)

    # Iteration over the number of epochs specified
    for e in range(self.epochs):
      del_w = np.zeros(self.weights.shape)
      # Iteration over the data rows
      for x, y in zip(self.X.values, self.y):
        # Estimating linear equation
        h = np.dot(x, self.weights)

        # Estimate prediction probabilities
        output = self.sigmoid(x=h)

        # Estimate the error term i.e. a gradient w.r.t. weights
        # Created with the help of the sigmoid_prime i.e. derivative of the sigmoid
        # In other words, (y-yhat) * sigmoid_prime(h)
        error_term = (y-output) * output * (1-output)

        # Calculate the change in weights for this sample and add it to the total weight change
        del_w += error_term * x

      # Update weights using the learning rate and the average change in weights
      self.weights += self.learnrate * del_w / self.n_records

      # Printing out the mean square error on the training set
      # Estimating loss at every 100th epoch; Feel free to change the setting
      if e % (self.epochs / 10) == 0:
        out = self.sigmoid(x=np.dot(X, self.weights))
        loss = np.mean((out - y) ** 2)
        if self.last_loss and self.last_loss < loss:
          print("Train loss: ", loss, "  WARNING - Loss Increasing")
        else:
          print("Train loss: ", loss)
        self.last_loss = loss
        
  def predict(self, X):
    tes_out = self.sigmoid(x=np.dot(X, self.weights))
    predictions = tes_out > 0.5
    return predictions.astype(int)
```

**<h3>Note:</h3>**

- As an alternative you can also use open source library such as <a href="https://pypi.org/project/gradient-descent/">**gradient-descent**</a> and refer the <a href="https://github.com/DanielDaCosta/gradient-descent">**documentation**</a>.

- The above class is almost similar as present in gradient-descent but with several more functions that you can experiemnt with.

## **8.4 Mini-batch Gradient Descent**

```
%%time

# Randomize data points
X_train, y_train = shuffle(X_train, y_train)

# Setting minibatch size of 2^7; Generally used 2^val
minibatch_size = 128

for i in range(0, X_train.shape[0], minibatch_size):
  # Get pair of (X, y) of the current minibatch/chunk
  X_train_mini = X_train[i:i + minibatch_size]
  y_train_mini = y_train[i:i + minibatch_size]

  minibatch_clf = SGDClassifier(loss='log', n_jobs=-1, random_state=42, class_weight='balanced')
  minibatch_clf.fit(X_train_mini, y_train_mini)

print('Training Mini-batch success!')
```
